```python
gender=vd['Gender'].unique()
approved,not_approved=[],[]
for i in gender:
    con1=(vd['Gender']==i)
    con2=(vd['Loan_Status']=='Y')
    con3=(vd['Loan_Status']=='N')
    count_approved=len(vd[con1&con2])
    count_not_approved=len(vd[con1&con3])
    approved.append(count_approved)
    not_approved.append(count_not_approved)
approved,not_approved
```
